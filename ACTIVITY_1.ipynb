```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Example data (Sine Function)
x = np.linspace(0, 10, 100)
y = np.sin(x)  

# Create the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, label='sin(x)', color='blue')
plt.title('Simple Graph Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()

#COVID19 Dataset
df = pd.read_csv('country_wise_latest.csv')

x = df['WHO Region']
y = df['Active']


plt.figure(figsize=(8, 5))
plt.bar(x, y, label='Data from CSV', color='green')
plt.title('Graph from CSV Data')
plt.xlabel('WJ')
plt.ylabel('Y-axis')
plt.grid(True)
plt.legend()
plt.show()
```
